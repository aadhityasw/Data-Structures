# Krushal Minimum Spanning Tree

```python
def get_Input() :
    # Gets the graph from the user.
    graph = {}
    n = int(input('Enter the number of edges in the graph :- '))
    print('Enter the source, destination and the weight of the various edges seperated by a space :')
    for i in range(n) :
        inp = input('Enter the details of the edge :- ').split(' ')
        if inp[0] not in graph :
            graph[inp[0]] = []
        graph[inp[0]].append((inp[1], int(inp[2])))
        if inp[1] not in graph :
            graph[inp[1]] = []
    return graph

def sort_Edges(graph) :
    # Sorts the edges in order of its weights.
    edges = []
    # Stores source, destination and weight of all edges in the graph.
    for u in graph :
        for v, w in graph[u] :
            for s in edges :
                if s[2] > w :
                    i = edges.index(s)
                    edges.insert(i, [u, v, w])
                    break
            else :
                edges.append([u, v, w])
    return edges

# A Union Cycle detection strategy with ranks and path compression.
class Nodes :
    def __init__(self, parent, rank) :
        self.parent = parent
        self.rank = rank
        return

def find(nodeSet, node) :
    if nodeSet[node].parent != node :
        nodeSet[node].parent = find(nodeSet, nodeSet[node].parent)
    return nodeSet[node].parent

def union(nodeSet, x, y) :
    # To attach the tree with a smaller rank with the one with a larger rank.
    if nodeSet[x].rank > nodeSet[y].rank :
        nodeSet[y].parent = x
    elif nodeSet[y].rank > nodeSet[x].rank :
        nodeSet[x].parent = y
    else :
        nodeSet[x].parent = y
        nodeSet[y].rank += 1
    return nodeSet

def detect_Cycle(edge, nodeSet) :
    # Returns True if there exists a cycle in the graph as the new edge is added and False otherwise.
    x = find(nodeSet, edge[0])
    y = find(nodeSet, edge[1])
    if x == y :
        return True
    else :
        nodeSet = union(nodeSet, x, y)
        return False

def add_Edge(edge, graphs) :
    # Adds an edge to a graph if it is not already present in it.
    destination = [edge[1], edge[2]]
    if destination not in graphs[edge[0]] :
        graphs[edge[0]].append(destination)
    return graphs

def remove_Edge(edge, graphs) :
    # Removes an edge from a graph if it is present in it.
    destination = [edge[1], edge[2]]
    if destination in graphs[edge[0]] :
        graphs[edge[0]].remove(destination)
    return graphs

def Krushal_Minimum_Spanning_Tree(edges, graphMST, V, frontier) :
    # The terminating condition is that the number of edges in the MST is one less than the number of vertices in the graph.
    itr = 0
    edgeCount = 0
    nodeSet = {}
    for v in graph :
        nodeSet[v] = Nodes(v, 0)
    while (itr < len(edges)) and (edgeCount < V) :
        graphMST = add_Edge(edges[itr], graphMST)
        if detect_Cycle(edges[itr], nodeSet) :
            graphMST = remove_Edge(edges[itr], graphMST)
        else :
            edgeCount += 1
        itr += 1
    return graphMST

def print_MST(graphMST) :
    print('The edges of the Minimum Spanning tree along with their respective weights are as follows : ')
    print('Source --> Destination ===> Weight')
    for u in graphMST.keys() :
        for v, w in graphMST[u] :
            print(u, '-->', v, '===>', w)

graph = get_Input()
#graph = graph_1
edges = sort_Edges(graph)
graphMST = {}
frontier = {}
for v in graph.keys() :
    graphMST[v] = []
    frontier[v] = 0
V = len(graph.keys())
graphMST = Krushal_Minimum_Spanning_Tree(edges, graphMST, V, frontier)
print_MST(graphMST)
```

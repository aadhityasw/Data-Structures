# Binary Search Tree

### The BST sort is just the in-order traversal of the tree.

```python
class Node :
    def __init__(self, key) :
        self.left = None
        self.right = None
        self.val = key
        self.parent = None

def insert(root, key) :
    if root is None :
        root = Node(key)
    else :
        if root.val < key :
            if root.right is None :
                temp = Node(key)
                root.right = temp
                temp.parent = root
            else :
                root = insert(root.right, key)
        else :
            if root.left is None :
                temp = Node(key)
                root.left = temp
                temp.parent = root
            else :
                root = insert(root.left, key)
    return(root)
'''
def delete(root, key) :
    r = find(root, key)
    if r is None :
        print('Element not found.')
    else :
        if r.left is None :
            if r.parent.left is r :
                r.parent.left = r.right
                r.right.parent = r.parent
                print('Element Deleted')
                return()
            elif r.parent.right is r :
                r.parent.right = r.right
                r.right.parent = r.parent
                print('Element Deleted')
                return()
        elif r.right is None :
            if r.parent.left is r :
                r.parent.left = r.left
                r.left.parent = r.parent
                print('Element Deleted')
                return()
            elif r.parent.right is r :
                r.parent.right = r.left
                r.left.parent = r.parent
                print('Element Deleted')
                return()
        else :
            temp = successor_node(r)
            r.val, temp.val = temp.val, r.val
            delete(r, key)
'''
# Another Way of performing the delete function.
def delete(r, key):
    root = find(r, key)
    if root is None :
        print('Element not found.')
        return()
    if root.left is None or root.right is None:
        if root is root.parent.left:
            root.parent.left = root.left or root.right
            if root.parent.left is not None:
                root.parent.left.parent = root.parent
        else:
            root.parent.right = root.left or root.right
            if root.parent.right is not None:
                root.parent.right.parent = root.parent 
        print('Element Deleted')
    else:
        s = successor_node(root)
        root.val, s.val = s.val, root.val
        return delete(root, s.val)

def check(root, lokey, hikey):
    f = True
    g = True
    if lokey is not None and root.val < lokey:
        return(False)
    if hikey is not None and root.val > hikey:
        return(False)
    if root.left is not None:
        if root.left.parent is not root:
            return(False)
        f = check(root.left, lokey, root.val)
    if root.right is not None:
        if root.right.parent is not root:
            return(False)
        g = check(root.right, root.val, hikey)
    if f == False or g == False :
        return(False)
    return(True)

def find(root, key) :
    if root is None :
        return(root)
    if root.val > key :
        return(find(root.left, key))
    elif root.val < key :
        return(find(root.right, key))
    elif root.val == key :
        return(root)

def minimum_node(root) :
    min_node = root
    while min_node.left is not None :
        min_node = min_node.left
    return(min_node)

def successor_node(node) :
    if node.right is not None :
        return(minimum_node(node.right))
    while node.parent is not None and node.parent.right is node :
        node = node.parent
    return(node.parent)

def inorder(root) :
    if root is not None :
        inorder(root.left)
        print(root.val)
        inorder(root.right)

def preorder(root) :
    if root is not None :
        print(root.val)
        preorder(root.left)
        preorder(root.right)

def postorder(root) :
    if root is not None :
        postorder(root.left)
        postorder(root.right)
        print(root.val)

def back_to_root(root) :
    while root.parent is not None :
        root = root.parent
    return(root)

root = None
s = input()
s = s.split(' ')
for i in range(len(s)) :
    root = insert(root, int(s[i]))
root = back_to_root(root)

n = int(input('Enter the element to be deleted :- '))
delete(root, n)
root = back_to_root(root)

print('In order Traversal')
inorder(root)

print('Pre order Traversal')
preorder(root)

print('Post order Traversal')
postorder(root)

n = int(input('Enter the element to be found in the tree :- '))
if find(root, n) is not None :
    print('Result : ', find(root, n).val)
else :
    print('Not Found')

print('Result of checking if the tree is actually a BST or not!!')
a, b = map(int, input('Enter the numbers for the check constraint :- ').split(' '))
print('Result : ',check(root, a, b))
```
